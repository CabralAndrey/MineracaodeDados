### Calculando e Plotando a média obtida para cada uma das linguagens de programação

```python
#Obtendo e plotando média de habilidades por linguagem
languages_names = list(languages_df.columns)

languages_mean_dict = {}
for name in languages_names:
    language_mean = languages_df[name].to_numpy().mean()
    languages_mean_dict[name] = language_mean
    
my_df = pd.DataFrame(languages_mean_dict.items())

sns.set_theme(rc={'figure.figsize':(11.7,8.27)})


ax = sns.barplot(x=0, y=1, data=my_df)
ax.set(xlabel='Linguagens', ylabel='Média', title='Média de Habilidade por Linguagem de Programação')
```

```python
#Discretização de habilidades
number_bins = 3
languages_disct_df = languages_df.copy()

for name in languages_names:
    languages_disct_df[name] = pd.qcut(languages_disct_df[name], number_bins, ["Baixa", "Média", "Alta"])
    
languages_disct_df.head()
```

```python
plot_df = pd.DataFrame(languages_disct_df)

# Contagem das ocorrências de cada categoria para cada linguagem
count_df = plot_df.apply(lambda x: x.value_counts()).T.fillna(0)
count_df.head()


# Plotagem do gráfico de barras
fig, ax = plt.subplots(figsize=(12, 8))

categories = ['Alta', 'Média', 'Baixa']
bar_width = 0.2
index = np.arange(len(count_df))

for i, category in enumerate(categories):
    ax.bar(index + i * bar_width, count_df[category], bar_width, label=category)

# Configurações do gráfico
ax.set_xlabel('Linguagem de Programação')
ax.set_ylabel('Ocorrencias')
ax.set_title('Nível de Habilidade por Linguagem de Programação')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(count_df.index, rotation=45)
ax.legend()

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```

```python
#COMO OS DADOS ESTÃO CATEGORIZADOS EM BAIXA MEDIA E ALTA PRECISAMOS TRANSFORMAR EM NUMEROS

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
#AGORA QUE ESTÁ FEITO A TRANSFORMAÇÃO DOS DADOS É SÓ APLICAR O COMANDO PADRÃO DIVIDINDO O CONJUNTO EM 80% TREINAMENTO E 20% TESTE
#from sklearn.cross_validation import train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#TREINANDO O KNN
from sklearn.neighbors import KNeighborsClassifier

# RODANDO KNN COM 3 VIZINHOS (NÃO DEU TEMPO DE RODAR COM MAIS) 
knn_model = KNeighborsClassifier(n_neighbors=3)# fazer um tunning talvez???#5

# TRAINANDO COM OS DADOS QUE FORAM SEPARADOS PARA TREINO ALI EM CIMA
knn_model.fit(X_train, y_train)
```

```python
#AVALIANDO DESEMPENHO COM O OS DADOS DE TESTE QUE FORAM GERADOS

from sklearn.metrics import accuracy_score
# VERIFICANDO PREVISÕES
y_pred = knn_model.predict(X_test)

# VERIFICANDO OS ACERTOS DO NOSSO KNN
accuracy = accuracy_score(y_test, y_pred)
print(f"Acurácia do modelo KNN: {accuracy:.2f}")
```

```python
#FICOU UMA MERDA...
#TESTANDO COM 5 VIZINHOS.
knn_model2 = KNeighborsClassifier(n_neighbors=5)

# TRAINANDO COM OS DADOS QUE FORAM SEPARADOS PARA TREINO ALI EM CIMA
knn_model2.fit(X_train, y_train)
```

```python
#TENTANDO UM TUNNING # ESTE TUNNING PEGUEI DO MANUAL DO SKLEARN
from sklearn.model_selection import cross_val_score
# creating list of K for KNN
k_list = list(range(1,50,2))
# creating list of cv scores
cv_scores = []

# perform 10-fold cross validation
for k in k_list:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())
```
